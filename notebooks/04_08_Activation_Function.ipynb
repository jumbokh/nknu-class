## 激勵函數(Activation Function)

```python
import tensorflow as tf
from tensorflow.keras import activations
import numpy as np
import matplotlib.pyplot as plt
```

## ReLU(Rectified linear unit)

```python
# 設定 x = -10, -9, ..., 10 測試
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)

# ReLU
y = activations.relu(x_tf).numpy()

# 繪圖
plt.plot(x, y)
plt.show()
```

## ReLU 參數

```python
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)
y = activations.relu(x_tf, threshold=5).numpy()

plt.plot(x, y)
plt.show()
```

```python
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)
y = activations.relu(x_tf, max_value=5).numpy()

plt.plot(x, y)
plt.show()
```

```python
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)
y = activations.relu(x_tf, alpha=0.5).numpy()
print(y)

plt.plot(x, y)
plt.show()
```

## Sigmoid

```python
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)

# sigmoid
y = activations.sigmoid(x_tf).numpy()

# 模糊地帶
plt.axvline(-4, color='r')
plt.axvline(4, color='r')

plt.plot(x, y)
plt.show()
```

## 使用NumPy計算Sigmoid

```python
x= np.linspace(-10, 10, 21)
y = 1 / (1 + np.e ** (-x))

plt.plot(x, y)
plt.show()
```

## tanh

```python
x= np.linspace(-10, 10, 21)
x_tf = tf.constant(x, dtype = tf.float32)

# tanh
y = activations.tanh(x_tf).numpy()

# 模糊地帶
plt.axvline(-3, color='r')
plt.axvline(3, color='r')

plt.plot(x, y)
plt.show()
```

## Softmax

```python
# activations.softmax 輸入需為 2 維資料，設定 x 為均勻分配，轉換後每一列加總為 1
x = np.random.uniform(1, 10, 40).reshape(10, 4)
print('輸入：\n', x)
x_tf = tf.constant(x, dtype = tf.float32)

# Softmax
y = activations.softmax(x_tf).numpy()
print('加總：', np.round(np.sum(y, axis=1)))
```

## 使用NumPy計算Softmax

```python
# 設定 x = 1, 2, ..., 10 測試
x= np.random.uniform(1, 10, 40)

# Softmax
y = np.e ** (x) / tf.reduce_sum(np.e ** (x))
print(sum(y))
```
